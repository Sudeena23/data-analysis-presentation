#  **Line Plot of Opening Prices Over**

```python

# Make sure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort data by date (just in case)
df = df.sort_values(by='Date')

# Plotting the % Change over time
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Open'], color='blue', linewidth=1)
plt.title('Open', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Open')
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. shows how the NEPSE market opened each trading day from 2020 to 2025.

2. An upward trend suggests growing investor confidence, while downward movements may reflect uncertainty or market stress.

3. Sharp fluctuations could correspond with major national or global events (e.g., COVID-19, political changes).

# **Line Plot of Closing Prices Over Time**


```python

# Make sure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort data by date (just in case)
df = df.sort_values(by='Date')

# Plotting the % Change over time
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['Close'], color='blue', linewidth=1)
plt.title('Close', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Close')
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. Reflects the end-of-day market value, providing a better indication of long-term market direction.

2. Trends in closing prices help identify overall market growth, corrections, or downturns.

3. Comparing closing prices across time can reveal recovery periods or prolonged bearish phases.
